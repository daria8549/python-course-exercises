```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
        
class SortedLinkedList:
    def __init__(self):
        self.first = None
        
    def insert(self, value):
        node = Node(value)
        
        if self.first is None or self.first.value > value:
            node.next = self.first
            self.first = node
            return
        
        current = self.first
        while current.next is not None and current.next.value < value:
            current = current.next
            
        node.next = current.next
        current.next = node
        return
            
            
    def printList(self):
        result = []
        current = self.first
        while current is not None:
            result.append(str(current.value))
            current = current.next
        
        return ', '.join(result) + ','

```

```python
my_ll = SortedLinkedList()
l = [5, 8, 3, 2, 1, 4, 6, 9, 7]
for value in l:
    my_ll.insert(value)
my_ll.printList()
assert my_ll.printList() == "1, 2, 3, 4, 5, 6, 7, 8, 9,"
```

# Question 6:
 Implement the following classes and methods of a binary search tree:
 • The class BSTNode should have an integer value, and a right and left child (successor).
   When creating a new node, value should be passed in the constructor.
 • The class BinarySearchTree should provide the method level(value). 
   This method should return the level of the node containing value, or None if value is not in the tree.
 • The class BinarySearchTree should provide the recursive method contains(node, value).
   The method returns True if value is present in the tree, or False otherwise.
 Note: The recursive method add has already been implemented as a helper,
       which adds new entries to the tree.


```python
class BSTNode:
    def __init__(self,value):
        self.value = value
        self.right = None
        self.left = None
```

```python
class BinarySearchTree:
    def __init__(self):
        self.root = None
  
    def add(self, n, value):
        if self.root is None:
            self.root = BSTNode(value)
        else:
            if n is None:
                return BSTNode(value)
            if value < n.value:
                n.left = self.add(n.left, value)
            else:
                n.right = self.add(n.right, value)
        return n
  
    def contains(self, node, value):
        if self.root is None:
            return False
        elif (node.value == value):
            return True
        elif (value < node.value and node.left != None):
            return self.contains(node.left, value)
        elif (value > node.value and node.right != None):
            return self.contains(node.right, value)
        return False
  
    def level(self, value):
        if self.root is None or self.contains(self.root, value) is False:
            return None
        current = self.root
        level = 0
        while current is not None:
            if current.value == value:
                return level
            if value > current.value:
                current = current.right
            else:
                current = current.left
            
            level += 1
        return None
    
            
            
        
```

```python
# Example usage
bst = BinarySearchTree()
L = [10, 4, 18, 5, 12]
for i in L:
    bst.add(bst.root, i)

# Test cases
assert bst.level(10) == 0
assert bst.level(5) == 2
assert bst.level(7) == None
assert bst.contains(bst.root, 12) == True

bst2 = BinarySearchTree()
assert bst2.contains(bst2.root, 12) == False
```

```python
class File:
    def __init__(self, filename, content):
        self.filename = filename
        self.content = content
        
    def get_id(self):
        hc = 0
        for ch in self.filename:
            hc = 31 * hc + ord(ch)
        return hc

class HashTable:
    def __init__(self, length):
        self.max_length = length
        self.table = [None] * self.max_length
        
    def add_item(self, key, value):
        hashed_key = self.get_bucket(key)
        if self.table[hashed_key] is None:
            self.table[hashed_key] = []
        self.table[hashed_key].append(value)
        
        
    def get_item(self, key):
        hashed_key = self.get_bucket(key)
        if self.table[hashed_key] is None:
            return "not found"
        for f in self.table[hashed_key]:
            if f.get_id() == key:
                return f
        return "not found"
    
    def get_bucket(self, key):
        return key % self.max_length

# Example usage
file_table = HashTable(9)
f1 = File("readme.txt", "Hello World!")
f2 = File("license.txt", "Creative Commons Attribution 4.0 International")
file_table.add_item(f1.get_id(), f1)
file_table.add_item(f2.get_id(), f2)

# Find f1 in the table
result = file_table.get_item(f1.get_id())
assert result.content == "Hello World!"
```
