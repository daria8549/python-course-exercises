```python
class ListItem:
    def __init__(self, value):
        self.value = value
        self.next = None

class Queue:
    def __init__(self):
        self.first = None
        self.last = None

    def is_empty(self):
        return self.first is None

    def leave(self):
        if self.is_empty():
            return None
        
        current = self.first
        if self.first == self.last:  # If there's only one element
            self.first = self.last = None
            return current.value
        
        while current.next != self.last:
            current = current.next
        
        leaving_item = self.last
        self.last = current
        self.last.next = None
        return leaving_item.value

    def print_queue(self):
        p = self.first
        while p is not None:
            print(p.value, end=', ')
            p = p.next
        print()

    def enter(self, v):
        item = ListItem(v)
        if self.first is None: 
            self.first = self.last = item
        else:
            self.last.next = item
            self.last = item
            
            
    def replace(self, i, j):
        if i == j:
            return
        
        current = self.first
        length = 0
        while current is not None:
            length += 1
            current = current.next
        
        if i<0 or j<0 or i>=length or j>=length:
            return
        
        current = self.first
        node_i = None
        node_j = None
        counter = 0
        while current is not None:
            if counter == i:
                node_i = current
            if counter == j:
                node_j = current
            current = current.next
            counter += 1
            
        if node_i is not None and node_j is not None:
            node_i.value, node_j.value = node_j.value, node_i.value

```

```python
# Some test cases
q = Queue()
assert q.is_empty() == True
q.enter("head")
q.enter("shoulders")
q.enter("knee")
q.print_queue()  # Expected output: head, shoulders, knee,
assert q.is_empty() == False
#o = q.leave()
#assert o == "head"
q.print_queue()  # Expected output: shoulders, knee,
q.replace(0, 1)
q.print_queue()  # Expected output after swap: knee, shoulders,
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.is_XOR = False
        self.left = None
        self.right = None

class BoolTree:
    def __init__(self):
        self.root = None

    def _get_bool_value(self, n):
        if n.right is None and n.left is None:
            return n.value
        elif n.right is not None and n.left is not None: 
            n_left = self._get_bool_value(n.left) 
            n_right = self._get_bool_value(n.right)
            if n.is_XOR:
                return n_left != n_right
            else:
                if n_left is True and n_right is True:
                    return True
                return False
                

    def get_bool_value(self):
        return self._get_bool_value(self.root)

```

```python
# Creating nodes for testing
t1 = Node(True)
f1 = Node(False)
f2 = Node(False)

# Creating an AND node
l_and = Node(None)
l_and.left = t1
l_and.right = f1

# Creating an XOR node
l_xor = Node(None)
l_xor.is_XOR = True
l_xor.left = l_and
l_xor.right = f2

# Creating BoolTree instances for testing
bt0 = BoolTree()
bt0.root = t1  # Single true node

bt1 = BoolTree()
bt1.root = l_and  # AND operation node

bt2 = BoolTree()
bt2.root = l_xor  # XOR operation node


# Assertions to verify functionality
assert bt0.get_bool_value() == True
assert bt1.get_bool_value() == False
assert bt2.get_bool_value() == False

print("Result:", bt2.get_bool_value())  # Expected output: False
```
