```python
class Node:
    def __init__(self,name):
        self.name = name
        self.right = None
        self.left = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def is_leaf(self, n, name):
        if n is None:
            return False
        
        if n.name == name:
            if n.left is None and n.right is None:
                return True
            else:
                return False
        
        if  name < n.name:
            return self.is_leaf(n.left,name)
        else:
            return self.is_leaf(n.right,name)
        
    
    def height(self, node):
        if (node == None):
            return 0
        else:
            leftHeight = self.height(node.left)
            rightHeight = self.height(node.right)

            if (leftHeight > rightHeight):
                return leftHeight + 1
            else:
                return rightHeight + 1

    def add(self, n, name):
        if self.root == None:
            self.root = Node(name)
        else:
            if n == None:
                return Node(name)
            if name < n.name:
                n.left = self.add(n.left, name)
            else:
                n.right = self.add(n.right, name)
        return n


```

```python
bt = BinaryTree()
bt.add(bt.root, "Bob")
bt.add(bt.root, "Charly")
bt.add(bt.root, "Alice")
bt.add(bt.root, "Dora")
assert bt.height(bt.root) == 3
assert bt.is_leaf(bt.root, "Charly") == False
assert bt.is_leaf(bt.root, "Alice") == True
```

```python
class Element:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.first = None
        self.last = None

    def is_empty(self):
        return self.first == None

    def insert(self, v):
        temp = Element(v)
        if self.first == None:
            # empty list
            self.first = temp
            self.last = temp
        else:
            temp.next = self.first
            self.first = temp

    def print_list(self):
        p = self.first
        while p is not None:
            print(p.value, end=', ')
            p = p.next
        print()

    def contains(self, value):
        if self.first is None:
            return False
        current = self.first
        while current is not None:
            if current.value == value:
                return True
            current = current.next 
        return False

    def remove(self, i):
        if self.first is None or i<0:
            return False
        if i == 0:
            value = self.first.value
            self.first = self.first.next
            return value
        
        current = self.first
        previous = None
        index = 0
        
        while current is not None and index < i:
            previous = current
            current = current.next 
            index += 1
        
        if current is None:
            return None
        
        value = current.value
        previous.next = current.next
        
        if current == self.last:
            self.last = previous
            
        return value
            
```

```python
# Create a new linked list and insert some elements
ll = LinkedList()
ll.insert(10)
ll.insert(20)
ll.insert(30)

# Test contains method
print("Contains 20:", ll.contains(20))  # Expected: True
print("Contains 40:", ll.contains(40))  # Expected: False

```

```python
# Test remove method
print("Remove element at index 1 (should be 20):", ll.remove(1))  # Expected: 20
ll.print_list()  # Expected: 30, 10,

print("Remove element at index 0 (should be 30):", ll.remove(0))  # Expected: 30
ll.print_list()  # Expected: 10,

print("Remove element at index 5 (out of range):", ll.remove(5))  # Expected: None
ll.print_list()  # Expected: 10,
```
