## Exercise 1 - Simple math

1. Create the variables `a` and `b` and assign to them the values `252` and `151`, respectively.
2. Use the basic math operators (`+`, `-`, `*`, `/`) to calculate the sum, difference, product and quotient of `a` and `b`. Save each result in a separate variable.
3. Use the integer divison operator `//` to divide `a` by `b`. What is the difference to the regular divison using `/`?
4. Use `print` to output the results.

```python
a = 252
b = 151

ab_sum = a + b
ab_diff = a - b
ab_prod = a * b
ab_quot = a / b # The regular division always yields a floating-point number
ab_int_quot = a // b # The integer division cuts off any decimal places

print('a + b =', ab_sum)
print('a - b =', ab_diff)
print('a * b =', ab_prod)
print('a / b =', ab_quot)
print('a // b =', ab_int_quot)
```

## Exercise 4 - FizzBuzz



FizzBuzz is a popular coding exercise that started as a way to quickly assess a person's programming skills in a job interview. The idea is simple, but there are multiple ways to solve the task. We want to go over the numbers 1 through 20 and output something based on the following rules:

- If the number is divisible by 3, output "Fizz".
- If the number is divisible by 5, output "Buzz".
- If the number is divisible by both 3 and 5, output "FizzBuzz".
- If the number is neither divisible by 3 nor by 5, output the number itself.

The picture above illustrates what this might look like.

```python
for i in range(1, 21): # range(1, 21) starts at 1 and ends *before* 21, meaning 20 is the last number
    if i % 3 == 0 and i % 5 == 0: # i % n == 0 asks if i divided by n has a remainder of 0
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

## Exercise 5 - Loops

Loops allow repeating an action over and over. This can be used to, for example, walk through a list and look at each element. Below is a definition for a list called `numbers`. Use a loop or multiple loops to determine the maximum, minimum and average of the numbers.

**Note:** Python has built-in functions that make this trivial. The goal is to **not** use them and instead practice using loops.

```python
numbers = [7, 9, 5, 4, 2, 6, 3, 8, 0, 1]

num_min = numbers[0] # We start by assuming the first number is the smallest
num_max = numbers[0] # We start by assuming the first number is the largest
num_sum = 0 # The initial sum is 0

for i in numbers:
    if i < num_min: # We check whether the loop has found a new smallest number
        num_min = i # If so, we update the smallest number
        
    if i > num_max: # We check whether the loop has found a new largest number
        num_max = i # If so, we update the largest number
        
    num_sum += i # We add the current number to the sum

print('min =', num_min)
print('max =', num_max)
print('sum =', num_sum)
```
